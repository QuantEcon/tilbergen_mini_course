# Time series data

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
```

```python
dowjones = pd.read_csv('dji.csv', usecols=['date', 'dji'])  # Data from Wharton Research Data Services
# https://wrds-web.wharton.upenn.edu/wrds/ds/djones/djd/index.cfm?navId=183
```

```python
dowjones.head()
```

```python
dowjones['date'] = pd.to_datetime(dowjones['date'], format='%d%b%Y')
dowjones.head()
```

```python
dowjones.set_index('date', inplace=True)
dowjones.head()
```

```python
dowjones.plot(figsize=(10, 8))
plt.show()
```

```python
dowjones['2000':].plot(figsize=(10, 8))
plt.show()
```

```python
dowjones_yearly = dowjones.resample('Y').first()
dowjones_yearly.plot(figsize=(10, 8))
plt.show()
```

```python
dowjones_pc = dowjones.pct_change()
dowjones_pc.plot(figsize=(15, 5))
plt.show()
```

```python
dowjones_pc = dowjones.pct_change()
mean = dowjones_pc.mean()[0]
std = dowjones_pc.std()[0]
dowjones_pc.plot(figsize=(15, 10))
plt.axhline(mean, c='black', ls='--')
plt.axhline(mean + 2 * std, c='black', ls='--', alpha=0.6)
plt.axhline(mean - 2 * std, c='black', ls='--', alpha=0.6)
plt.show()
```
