### While loops

```python
ts_length = 100
ϵ_values = []
i = 0

while i < ts_length:
    e = np.random.randn()
    ϵ_values.append(e)
    i = i + 1
    
plt.plot(ϵ_values, 'b-')
plt.show()
```

With explicit branching:

```python
def g2(x, y):
    """
    Does the same thing as g
    
    """
    if x > y:
        mx = x
    else:
        mx = y
        
    return 2 * mx
        
```

```python
g2(3, 4)
```

Let's write a function for populating a Python list with draws from a standard normal distribution:

```python
def generate_data(n):
    ϵ_values = []
    for i in range(n):
        e = np.random.randn()
        ϵ_values.append(e)
    return ϵ_values
```

```python
data = generate_data(100)
```

```python
data[0:4]
```

```python
plt.plot(data)
plt.show()
```

A fancier function:

```python
def generate_data(n, generator_type):
    ϵ_values = []
    for i in range(n):
        if generator_type == 'U':
            e = np.random.uniform(0, 1)
        else:
            e = np.random.randn()
        ϵ_values.append(e)
    return ϵ_values
```

```python
data = generate_data(100, 'U')
```

```python
data[0:4]
```

```python
plt.plot(data)
plt.show()
```

Here's a nicer function that does the same thing:

```python
def generate_data(n, generator_type):
    ϵ_values = []
    for i in range(n):
        e = generator_type()
        ϵ_values.append(e)
    return ϵ_values
```

Note: we're passing a function to a function

```python
data = generate_data(100, np.random.uniform)
plt.plot(data)
plt.show()
```

Generate and plot a time series of length 500 from the AR(1) model

$$ X_{t+1} = \alpha X_t + \beta + \sigma \xi_{t+1} $$
    
when

* $\{\xi_t\}$ is IID and standard normal
* $\alpha = 0.8$, $\beta = 1.0$ and $\sigma = 0.2$

### Solutions

#### Exercise 1

Here's one approach, using list comprehensions:

```python
S = 0.0
S_values = []
for t in range(500):
    S += np.random.randn()
    S_values.append(S)
    
plt.plot(S_values, label="cumulative sum")
plt.legend()
plt.show()
```

Here's an approach that more fully exploits the functionality of the NumPy library:

```python
plt.plot(np.cumsum(np.random.randn(500)))
plt.show()
```

#### Exercise 2

```python
X_vec = []
X = 1.0
α, β, σ = 0.8, 1.0, 0.2

for t in range(500):
    X_vec.append(X)
    X = α * X + β + σ * np.random.randn()
    
plt.plot(X_vec, label='AR(1) process')
plt.legend()
plt.show()
```
